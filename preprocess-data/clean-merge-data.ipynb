### Waste data

```python
waste_per_county = pd.read_csv("../data/calrecycle_waste_per_county.csv")
waste_per_county.head()
```

```python
total_waste_per_county = waste_per_county.groupby("County").agg("sum")[["Total Residential Tons"]].reset_index()
total_waste_per_county.head()
```

```python
len(total_waste_per_county.County.unique())
```

```python
list(total_waste_per_county.County.str.lower().str.title())
```

### Electricity Data

```python
electricity_per_county = pd.read_csv("../data/cec-electricity-by-county.csv")
electricity_per_county.head()
```

```python
year = "2014"
```

```python
residential_cond = electricity_per_county.Sector == "Residential"
residential_electricity = electricity_per_county.loc[residential_cond, ["County", year]]
residential_electricity.rename(columns={year: "Electricity Usage_" + year}, inplace=True)
residential_electricity.head()
```

```python
len(residential_electricity.County.unique())
```

---

### Population data

```python
population_per_county = pd.read_csv("../data/census-ca-county-populations.csv", skiprows=3, skipfooter=5)
population_per_county.rename(columns={"Unnamed: 0":"County"}, inplace=True)
population_per_county.drop([0], inplace=True) # drop CA population
population_per_county.reset_index(drop=True, inplace=True)
population_per_county.tail()
```

```python
for col in population_per_county.columns:
    population_per_county[col] = population_per_county[col].str.replace(r'[,\.]', '')
    if col != "County":
        population_per_county[col] = population_per_county[col].astype(float)

population_per_county["County"] = population_per_county.County.str.replace(" County California", "").str.upper()
```

```python
population_per_county.head()
```

```python
population_df = population_per_county[["County", year]]
population_df.rename(columns={year: "Population_" + year}, inplace=True)
population_df.head()
```

```python
len(population_per_county.County.unique())
```

---

### Merge

```python
complete_feature_df = total_waste_per_county.merge(residential_electricity)
complete_feature_df = complete_feature_df.merge(population_df)
```

```python
complete_feature_df.head()
```

```python
complete_feature_df.to_csv("../data/complete_feature_df.csv", index=False)
```
